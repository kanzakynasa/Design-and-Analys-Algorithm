```python
def add(list_a, list_b):
  result = []
  x=0
  for first, second in zip(list_a, list_b):
    result.append(first + second)
  for i in result:
    x+=i
  return x
a =  [1,2,3]
b =  [4,5,6]
print(add(a,b))
```

```python
def get_kali(seq):
  hasil = 1
  for x in seq:
    hasil *= x
  return hasil

print(get_kali([1,2,3]))
print(get_kali([3,4,12]))
print(get_kali([]))
```

```python
def get_bagi(div):
  if not div:
    return 0

  hasil = div[0]
  for x in div[1:]:
    hasil /= x
  return hasil

print(get_bagi([144,12,6]))
print(get_bagi([81,9,3]))
print(get_bagi([]))
```

```python
def get_Sum(myList):
  sum = 0
  for row in myList:
    for item in row:
        sum += item
  return sum

get_Sum([[1,2,5],[3,4,7]])
```

```python
def get_Bagi(myList):

    sum1 = sum(myList[0])

    sum2 = sum(myList[1])

    if sum2 == 0:
        return None
    return sum1 / sum2

print(get_Bagi([[144, 12], [6]]))
print(get_Bagi([[81, 9], [3]]))
```

```python
def get_Kurang(myList):
    sum1 = sum(myList[0])
    sum2 = sum(myList[1])
    return sum1 - sum2

print(get_Kurang([[1,2,5],[3,4,7]]))
print(get_Kurang([[10,20],[5,3]]))
```
